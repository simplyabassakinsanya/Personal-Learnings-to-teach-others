WHAT IS PYTHON?

- Python is a programming language.

- A language is simply how people interact with each other.

For instance, while I admit my love for the italian language, i can not interact successfully with an italian (assuming he/she/them/they speaks fluently), I can barely say 'Buongiorno'.

In the same vein, if you can not speak and understand the yoruba language, we can not communicate effectively.

This means that for verbal communication to be effective, we must understand each other. This is not different from a programming language. Yes, it is the same!

- Python is similar to the English, Yoruba, or the Italian languages. Python is the language of programmers. It is how programmers communicate with computers, using syntax (symbols, elements, operators and its arrangements), rules (commands), as well as instructions (recipes). This means the communication/language is readable and easily understood.
