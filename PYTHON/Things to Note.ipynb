# Starting with Python
- Learning python requires a hands-on approach. This means you must practice everyday. It only sticks through consistent practice and repetition.

- There are different versions of python, similar to the different versions of other programming languages. The version with the highest number is the newest and the most recent model.

- To know the current version of python, you have to visit python.org (this is visible on the landing page). As of 20 March 2026, 12.15pm, the latest version of python source code is 3.14.4 even though my current version is 3.13.9.

- This means there have been some minor improvements in my current version of python package. This is important to note because what is applicable in older versions might have been refined or redefined. Hence, the need to be abreast of current happenings in the python world. #stay updated.
