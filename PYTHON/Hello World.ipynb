# Naming Conventions in Python: Naming conventions are like the etiquette of coding. They are crucial for writing clear and readable codes. If you violate these naming conventions, your code will result in a syntaxerror
1. All variables must begin with an uppercase letter (Abass, X, Sex, Age) or lowercase letter (abass, sex, age, x) or an underscore (_abass, _sex, _age, _x). # Must start with a letter or underscore.

    Accepted: my_variable, _internal_var, name
    Rejected: 1st_variable (starts with a number)

2. A variable can not start with a number.

3. A variable can only contain alphanumeric characters and underscores. After the first character, variable names can contain letters (a-z, A-Z), numbers (0-9), and underscores (_). No spaces, hyphens (-), or other special characters are permitted.

    Accepted: user_name_2, total_count
    Rejected: user-name, total count, item#price

4. Case-Sensitive: Python is a case-sensitive language. This means myVariable and myvariable are treated as two distinct variables.

5. Cannot Be Python Keywords: You cannot use Python's reserved keywords (like if, else, for, while, class, def, import, return, etc.) as variable names. Doing so will also result in a SyntaxError.

    Accepted: my_if_statement
    Rejected: if = 10

# Best Practices in naming conventions in Python: 

Beyond the strict rules, the Python community adheres to a style guide called PEP 8 (Python Enhancement Proposal 8). Following PEP 8 makes your code consistent with the vast majority of Python projects, making it easier for others (and your future self) to read and understand.

1. Use snake_case for Variables and Functions: For variable names, PEP 8 recommends using lowercase letters with words separated by underscores. This is known as snake_case.

 - Recommended: first_name, total_price, is_active
 - Avoid: firstName (camelCase), TOTALPRICE (all caps, usually for constants)


2. Choose Descriptive Names: Variable names should be clear and descriptive, indicating their purpose or the data they hold. Avoid overly short or ambiguous names unless their context is immediately obvious (e.g., loop counters like i or j).

 - Recommended: customer_age, product_quantity
 - Avoid: ca, pq, x (unless x is a temporary variable in a very small scope)


3. Avoid Single-Letter Names (Generally): While allowed, single-letter names (except for specific cases like loop iterators or coordinates) can make code harder to understand.


4. Avoid Leading/Trailing Underscores for Regular Variables:

 - A single leading underscore (_variable): Conventionally indicates a variable intended for internal use within a module or class. It's a hint to other programmers not to access it directly, though Python doesn't strictly prevent it.

 - A double leading underscore (__variable): Used for name mangling in classes to avoid naming conflicts in subclasses.

 - A single trailing underscore (variable_): Used to avoid conflicts with Python keywords (e.g., class_ instead of class).


5. Avoid Ambiguous Characters: Do not use the lowercase letter l, the uppercase letter O, or the uppercase letter I as single-character variable names. In some fonts, these can be easily confused with the number 1 or 0.

# Printing a variable: In printing a variable, all you have to do is use the print function. The print function is denoted using "print()". You will the input the variable in the () / open '(' and close ')' parenthesis. See examples below. 

```python
Nigeria = "leadership problems" #The value is in parenthesis because it is a string. The variable name is Nigeria. The value of the variable is "leadership problems".
print(Nigeria) #This line of code will print the value of the variable Nigeria, which is "leadership problems".
```

```python
ambition = "seeking phd opportunities"
print(ambition)

salary = 1000000
print(salary)

Male = True
print(Male)
```
