# Python Libraries

- The strenght of python lies in its ecosystem of over 100,000 libraries that exist today. 

- The most essential ones include NumPy, Pandas, Matplotlib, Scikit-learn, etc. Each of these libraries serve a distinct purpose and covers domains like numerical computation, data analysis, visualization, AI/ML, web frameworks, automation for developers and programmers.

- Python libraries are a collection of modules/packages with pre-written codes. These pre-written codes helps save time, redundancy, and provides specialized functionality. For instance, NumPy, a short form for Numerical Python is efficient in handling multi-dimensional arrays, matrices, and linear algebra. Scikit-learn, a short form for "Scientific Toolkit for Machine Learning" is a machine learning library.

# IDEs VS Python Code Editor VS Python Interpreter.

- IDEs are Integrated Development Environments. Think of it as a big school bag for coding. This big school bag contains

***Notebook***: It has a place to write your python code (a code editor)
***Eraser***: It helps you fix mistakes (debugging tools).
***Calculator***: This is the interpreter or compiler, which runs your code and show results.
***Folders***: It helps you organize big projects with many files
***Teacher's help***: It gives hints, auto-completes words, and show errors early.

Example of an IDE is PyCharm

- The Code Editor is what enables you to type the code. Using the analogy of a big school bag for coding, the code editor is the notebook, where you type the code. It does not run codes, but mainly used for writing.

- Code Interpreter: This is the part that actually reads your code and makes the computer do what you wrote. Basically, the interpreter makes you run the code.
